```python
#generates dataset of problems that are duplicates

import sys
import re
import pandas as pd
import os
import sys
```

```python
path = '''/w/225/1/chess/tactics/'''
os.chdir(path)
filename = '''tactics_problem.csv'''
```

```python
FEN_set = set([])
duplicate_FEN_set = set([])
count = 0
FEN_dict = {}
attempt_count_dict = {}
pattern = r'''FEN "[^"]*'''
for chunk in pd.read_csv(filename, chunksize=10000): 
    for index, row in chunk.iterrows():
        
        #print(row)
        #sys.exit()
        problem = row["tactics_problem_id"]
        pgn = row["pgn"]
        attempts = row["attempt_count"]
        #print(pgn)
        FEN = re.findall(pattern, pgn)[0][6:]
        attempt_count_dict[problem] = attempts
        if FEN not in FEN_set:
            FEN_set.add(FEN)
            FEN_dict[FEN] = [problem]
        else:
            #print(problem, FEN)
            count = count + 1
            duplicate_FEN_set.add(FEN)
            FEN_dict[FEN].append(problem)
            
        
```

```python
#print(len(FEN_set))
#print(count)
#print(len(duplicate_FEN_set))
dupl_file = open("duplicate_FENs.csv", "w")
to_write = "FEN,ID1,ID2\n"
duplicate_FEN_set = list(duplicate_FEN_set)
for i in range(len(duplicate_FEN_set)):
    fen = duplicate_FEN_set[i]
    ids = FEN_dict[duplicate_FEN_set[i]]
    attempts = [int(attempt_count_dict[f]) for f in ids]
    attempts, ids = (list(t) for t in zip(*sorted(zip(attempts, ids))))
    attempts.reverse()
    ids.reverse()
    #print(attempts, ids)
    if attempts[1] < 1000:
        print(attempts)
    else:
        attempts = attempts[0:2]
        ids = ids[0:2]
        
        to_write = to_write + str(fen) + "," + str(ids[0]) + "," + str(ids[1]) + "\n"
dupl_file.write(to_write)
```
