```python
import pandas as pd
import numpy as np
```

```python
tag_data = pd.read_csv("/w/225/1/chess/tactics/tag_data_updated.csv")
```

```python
num_of_rows, num_of_cols = tag_data.shape
new_tag_data = pd.DataFrame(index = np.arange(num_of_rows), columns = ['Puzzle ID', 'Tag'])
new_tag_data['Puzzle ID'] = tag_data['tactics_problem_id']
new_tag_data
```

```python
temp = tag_data.drop(columns = ['tactics_problem_id','Unnamed: 0', 'average_seconds', 'move_count'])
temp
```

```python
tag_data
```

```python
count_true_tag = {}
key_to_tag = {}
ind = 0
for col in temp.columns:
    count_true_tag[ind] = len(temp[temp[col] == True])
    key_to_tag[ind] = col
    ind = ind + 1
```

```python
row,col = tag_data.shape
for i in range(row):
    tags_list = np.array(temp.iloc[i,:])
    indices = np.nonzero(tags_list)
    minimum = count_true_tag[indices[0][0]]
    key = indices[0][0]
    for j in range(1, len(indices)):
        count_true = count_true_tag[indices[0][j]]
        if count_true < minimum:
            key = indices[0][j]
            minimum = count_true
    new_tag_data.iloc[i,1] = key_to_tag[key]
```

```python
new_tag_data.to_csv('single_tag_data.csv')
```
